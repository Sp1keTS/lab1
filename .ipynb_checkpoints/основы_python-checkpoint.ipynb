#Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def aeiou (string):
    result = 0
    for letter in "aeiou":
        result += string.count(letter)
    return result


def main():
    print(aeiou("aeiou") == 5)
    print(aeiou("hello world!") == 3)


if __name__ == "__main__":
    main()
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def only_unique_letters(string):
    result = True
    for letter in string:
        if string.count(letter) != 1:
            result = False
    return result


def main():
    print(only_unique_letters("hello world!") == False)
    print(only_unique_letters("aeiou") == True)

    
if __name__ == "__main__":
    main()
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
import unittest

def count_1_bytes(number):
    return bin(number).count('1')

class TestCount1Bytes(unittest.TestCase):
    
    def test(self):
        self.assertEqual(count_1_bytes(0), 0)   
        self.assertEqual(count_1_bytes(1), 1)
        self.assertEqual(count_1_bytes(1024), 1)
        self.assertEqual(count_1_bytes(255), 8)
        
if __name__ == '__main__':
    unittest.main()
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
import unittest

def magic(number):
    if number < 10:  
        return 0
    
    result = 0
    current = number
    while current >= 10:
        product = 1
        for digit in str(current):
            product *= int(digit)
        
        current = product
        result += 1
    
    return result

class TestCountBeforeSingle(unittest.TestCase):
    def test_from_cases(self):
        test_cases = [  
            (10, 1),    
            (25, 2),    
            (39, 3),    
            (4, 0),     
            (999, 4),   
            (77, 4),    
            (1, 0),     
            (0, 0), 
        ]
        
        for num, expected in test_cases:
            with self.subTest(num=num, expected=expected):
                self.assertEqual(magic(num), expected)

if __name__ == "__main__":
    unittest.main()
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
import unittest
import math

def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError()
    
    squared_errors = [(p - t) ** 2 for p, t in zip(pred, true)]
    return math.sqrt(sum(squared_errors) / len(pred))

class TestMSE(unittest.TestCase):
    def test_mse_cases(self):
        test_cases = [
            ([1, 2, 3], [1, 2, 3], 0.0),
            ([1, 2, 3], [4, 5, 6], 3.0),
            ([0, 0, 0], [1, 1, 1], 1.0),
            ([1, 3, 5], [2, 4, 6], 1.0),
            ([2, 4, 6], [1, 3, 5], 1.0),
        ]
        
        for pred, true, expected in test_cases:
            self.assertAlmostEqual(mse(pred, true), expected)

if __name__ == "__main__":
    unittest.main()
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
import unittest
def prime_factors(n):
    if n == 1:
        return "(1)"
    
    result = []
    divisor = 2
    
    while divisor * divisor <= n:
        count = 0
        while n % divisor == 0:
            count += 1
            n //= divisor
        if count > 0:
            if count == 1:
                result.append(f"({divisor})")
            else:
                result.append(f"({divisor}**{count})")
        divisor += 1
    
    if n > 1:
        result.append(f"({n})")
    
    return "".join(result)

class TestPrimeFactors(unittest.TestCase):
    def test_prime_factors(self):
        test_cases = [
            (1, "(1)"),
            (2, "(2)"),
            (4, "(2**2)"),
            (6, "(2)(3)"),
            (12, "(2**2)(3)"),
            (17, "(17)"),
            (100, "(2**2)(5**2)"),
            (30, "(2)(3)(5)"),
        ]
        
        for num, res in test_cases:
            self.assertEqual(prime_factors(num), res)

if __name__ == "__main__":
    unittest.main()

```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
import unittest

def pyramid(number):
    k = 1
    total = 0
    
    while total < number:
        total += k * k
        if total == number:
            return k
        k += 1
    
    return "It is impossible"

class TestPyramid(unittest.TestCase):
    def test_pyramid(self):
        test_cases = [
            (1, 1),
            (5, 2),
            (14, 3),
            (30, 4),
            (55, 5),
            (2, "It is impossible"),
            (7, "It is impossible"),
            (25, "It is impossible"),
            (100, "It is impossible"),
        ]
        
        for num, res in test_cases:
            self.assertEqual(pyramid(num), res)

if __name__ == "__main__":
    unittest.main()
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
import unittest

def is_balanced(number):
    digits = [int(d) for d in str(number)]
    n = len(digits)
    
    if n <= 2:
        return True
    
    if n % 2 == 1:  
        mid = n // 2
        left_sum = sum(digits[:mid])
        right_sum = sum(digits[mid + 1:])
    else:          
        mid = n // 2
        left_sum = sum(digits[:mid - 1])
        right_sum = sum(digits[mid + 1:])
    
    return left_sum == right_sum

class TestBalancedNumber(unittest.TestCase):
    def test_balanced(self):
        test_cases = [
            (12321, True),
            (123321, True),
            (12345, False),
            (123456, False),
            (11, True),
            (121, True),
            (1221, True),
            (1234, False),
            (1, True),
            (1234221, False),
        ]
        
        for num, res in test_cases:
            self.assertEqual(is_balanced(num), res)

if __name__ == "__main__":
    unittest.main()

```
